```python
import numpy as np
import random
import matplotlib.pyplot as plt

# Constants
SCREEN_WIDTH = 1900
SCREEN_HEIGHT = 1200
STREET_WIDTH = 10  # Uniform width of the street
NUM_STREETS = 5

# Create an empty screen (0 for no color, 1 for street)
screen = np.zeros((SCREEN_HEIGHT, SCREEN_WIDTH), dtype=int)

# Function to draw a street using Bézier curves
def draw_street(screen, points, street_width, color=1):
    for t in np.linspace(0, 1, 500):
        x = int((1 - t)**2 * points[0][0] + 2 * (1 - t) * t * points[1][0] + t**2 * points[2][0])
        y = int((1 - t)**2 * points[0][1] + 2 * (1 - t) * t * points[1][1] + t**2 * points[2][1])
        for i in range(-street_width // 2, street_width // 2):
            for j in range(-street_width // 2, street_width // 2):
                if 0 <= x + i < SCREEN_WIDTH and 0 <= y + j < SCREEN_HEIGHT:
                    screen[y + j][x + i] = color

# Function to generate random connected streets
def generate_random_connected_streets(screen, num_streets, street_width):
    # Start at a random point on the screen
    start_x = random.randint(100, SCREEN_WIDTH - 100)
    start_y = random.randint(100, SCREEN_HEIGHT - 100)
    
    for _ in range(num_streets):
        # Randomly choose the next control point and end point
        control_x = random.randint(max(0, start_x - 200), min(SCREEN_WIDTH, start_x + 200))
        control_y = random.randint(max(0, start_y - 200), min(SCREEN_HEIGHT, start_y + 200))
        
        end_x = random.randint(max(0, control_x - 200), min(SCREEN_WIDTH, control_x + 200))
        end_y = random.randint(max(0, control_y - 200), min(SCREEN_HEIGHT, control_y + 200))
        
        # Define points for the Bézier curve (start, control, end)
        points = [(start_x, start_y), (control_x, control_y), (end_x, end_y)]
        
        # Draw the street
        draw_street(screen, points, street_width)
        
        # The next street starts where the previous one ended
        start_x, start_y = end_x, end_y

# Function to display the screen
def display_screen(screen):
    plt.imshow(screen, cmap='gray')
    plt.show()

# Generate and display 5 connected streets
generate_random_connected_streets(screen, NUM_STREETS, STREET_WIDTH)
# display_screen(screen)
```

```python
import numpy as np

# Constants for screen size
SCREEN_WIDTH = 1900
SCREEN_HEIGHT = 1200

# Function to draw a point on the screen with a given width and color
def draw_point(screen, x, y, width, color=1):
    for i in range(-width // 2, width // 2):
        for j in range(-width // 2, width // 2):
            if 0 <= x + i < SCREEN_WIDTH and 0 <= y + j < SCREEN_HEIGHT:
                screen[y + j][x + i] = color
    return screen

# Function to draw a line between two points with a given width
def draw_line(screen, p1, p2, width, color=3):
    width = int(width)
    if p1[0] == p2[0]:  # Vertical line case
        for y in np.linspace(p1[1], p2[1], 500):
            screen = draw_point(screen, p1[0], int(y), width, color)
    else:
        a = (p1[1] - p2[1]) / (p1[0] - p2[0])  # Slope of the line
        b = p1[1] - a * p1[0]  # Intercept of the line
        for x in np.linspace(p1[0], p2[0], 500):
            y = a * x + b
            screen = draw_point(screen, int(x), int(y), width, color)
    return screen

# Function to draw and fill a quadrilateral (shape with 4 points)
def draw_and_fill(screen, p1, p2, p3, p4, width=1, color=3):
    # Draw the border lines of the shape
    screen = draw_line(screen, p1, p2, width, color)
    screen = draw_line(screen, p2, p3, width, color)
    screen = draw_line(screen, p3, p4, width, color)
    screen = draw_line(screen, p4, p1, width, color)
    
    # Now fill the shape by finding the minimum and maximum x and y values
    min_x = int(min(p1[0], p2[0], p3[0], p4[0]))
    max_x = int(max(p1[0], p2[0], p3[0], p4[0]))
    min_y = int(min(p1[1], p2[1], p3[1], p4[1]))
    max_y = int(max(p1[1], p2[1], p3[1], p4[1]))
    
    # Fill the area between the borders
    for x in range(min_x, max_x + 1):
        for y in range(min_y, max_y + 1):
            if 0 <= x < SCREEN_WIDTH and 0 <= y < SCREEN_HEIGHT:
                screen = draw_point(screen, x, y, width, color)
    
    return screen

# Example usage:
# Create an empty screen (0 means no color)
screen_req = np.zeros((SCREEN_HEIGHT, SCREEN_WIDTH), dtype=int)

# Define four points (x, y)
p1 = (300, 400)
p2 = (600, 500)
p3 = (700, 800)
p4 = (400, 700)

# Draw and fill the region (streets) connecting these points
screen_req = draw_and_fill(screen_req, p1, p2, p3, p4, 5, color=2)

# Display the screen using matplotlib (for visualization)
import matplotlib.pyplot as plt

def display_screen(screen):
    plt.imshow(screen, cmap='gray')
    plt.show()

display_screen(screen)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Function to calculate parallel line points
def parallel_line(p1, p2, dist):
    x1, y1 = p1
    x2, y2 = p2

    # Calculate the direction vector between p1 and p2
    dx = x2 - x1
    dy = y2 - y1

    # Calculate the length of the direction vector (distance between points)
    line_length = math.sqrt(dx**2 + dy**2)

    # Normalize the perpendicular vector (unit normal vector)
    unit_perpendicular = (-dy / line_length, dx / line_length)

    # Create the new parallel points by moving the original points
    p1_parallel = (
        x1 + dist * unit_perpendicular[0], 
        y1 + dist * unit_perpendicular[1]
    )
    p2_parallel = (
        x2 + dist * unit_perpendicular[0], 
        y2 + dist * unit_perpendicular[1]
    )

    return p1_parallel, p2_parallel

# Example points and distance
p1 = (1, 2)
p2 = (2, 4)
distance = 2  # Distance to the parallel line

# Get the parallel line points
parallel_p1, parallel_p2 = parallel_line(p1, p2, distance)

# Plotting the original and parallel lines
plt.figure(figsize=(6,6))

# Original line
x_values = [p1[0], p2[0]]
y_values = [p1[1], p2[1]]
plt.plot(x_values, y_values, label="Original Line", color='blue')

# Parallel line
parallel_x_values = [parallel_p1[0], parallel_p2[0]]
parallel_y_values = [parallel_p1[1], parallel_p2[1]]
plt.plot(parallel_x_values, parallel_y_values, label="Parallel Line", color='red')

# Marking the points
plt.scatter([p1[0], p2[0]], [p1[1], p2[1]], color='blue', zorder=5)
plt.scatter([parallel_p1[0], parallel_p2[0]], [parallel_p1[1], parallel_p2[1]], color='red', zorder=5)

# Labels and title
plt.legend()
plt.title("Original Line and Parallel Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)

# Show plot
plt.show()
```
